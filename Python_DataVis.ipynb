#Numpy

###Creating arrays

```python
import numpy as np
a=np.array([1,2,3,4,5])
b=np.arange(1,6) #Creates [1,2,3,4,5]
c=np.zeros((2,3))  #2x3 matrix of zero
d=np.ones((3,2)) #3x2 martix of ones

print("Array a:",a)
print("Array b:",b)
print("2x3 Zeros:",c)
print("3x2 Ones:",d)
```

### Indexing & Slicing

```python
print("First Element of a:",a[0])
print("Last 2 Element of a:",a[-2:])
```

###Basic Arthmetic operations


```python
print("a+2:",a+2)
print("a*3:",a*3)
```

###Basic Statistics

```python
print("Square of a:",np.square(a))
print("mean of a:",np.mean(a))
print("Standard Deviation of a:",np.std(a))
```

###Reshaping Arrays

```python
matrix=np.arange(1,10).reshape((3,3))
print("Reshaped 3x3 matrix:")
matrix
```

###Transpone

```python
print("Transpose:")
matrix.T
```

###Boolean Indexing

```python
print("Element > 5:",matrix[matrix>5])
```

#Data PreProcessing in python

**Why Data Preprocessing?**

It can have missing  values,wrong formats,different scales or categories that need convesion

##Handling Missing Data


*  Human Error
*  Data Corruption
*  Not applicable field  





##Check for Missing Values

```python
import pandas as pd

df=pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
df.isnull().sum()
```

```python
df[df.isnull().any(axis=1)]
```

```python
#pd.set_option("display.max_rows",None)
df[df['Age'].isnull()]
```

```python
df.iloc[5]
```

```python
df.info()
```

##A) Drop  Rows or Columns

```python
#df.dropna(inplace=True) #drop rows within any missing values
#df.dropna(subset=['Age'],inplace=True)
df.drop(columns=['Cabin'],inplace=True)
```

##B) Fill with Static Value

```python
df['Embarked'].fillna('Unknown',inplace=True)
df['Age'].fillna(0,inplace=True)
```

##C) Fill With Mean,Median,Mode

```python
df['Age'].fillna(df['Age'].mean(),inplace=True)
df['Age'].fillna(df['Age'].median(),inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0],inplace=True)
```

## D) Forward/Backward Fill

```python
df.fillna(method='ffill',inplace=True)
df.fillna(method='bfill',inplace=True)
```

##Final Handling

```python
df.drop(columns=['Cabin'],inplace=True)
df['Embarked']=df['Embarked'].fillna(df['Embarked'].Mode()[0])
df['Age']=df['Age'].fillna(df.['Age'].mean())
```

```python
df.info()
```

##Encoding Categorial Variables

##A) Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
df['Sex'] = le.fit_transform(df['Sex']) #Male->1,Female->0
df
```

##B) One Hot Encoding

```python
df =pd.get_dummies(df ,columns=['Embarked'],drop_first=True)
```
